### Normalization Techniques
- Linear Scaling
- Z scaling
- Log scaling

| Normalization technique | Formula                                           | When to use                                                         |
|-------------------------|--------------------------------------------------|---------------------------------------------------------------------|
| Linear scaling          | x' = (x - xmin) / (xmax - xmin)                   | When the feature is uniformly distributed across a fixed range.     |
| Z-score scaling         | x' = (x - μ) / σ                                  | When the feature distribution does not contain extreme outliers.    |
| Log scaling             | x' = log(x)                                       | When the feature conforms to the power law.                         |
| Clipping                | If x > max, set x' = max <br> If x < min, set x' = min | When the feature contains extreme outliers.                         |
